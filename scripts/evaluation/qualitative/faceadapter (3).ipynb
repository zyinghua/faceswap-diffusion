```python
# inference script
%%writefile run_faceadapter_batch.py

import os
import sys
from PIL import Image
import shutil

def run_faceadapter(src_dir, tgt_dir, output_dir):
    """
    Run FaceAdapter on all image pairs and save results
    """
    # using existing infer.py script
    os.system(f"python infer.py -s {src_dir} -t {tgt_dir} -o {output_dir}")

if __name__ == '__main__':
    os.makedirs('test_images/source', exist_ok=True)
    os.makedirs('test_images/target', exist_ok=True)

    run_faceadapter(
        src_dir='test_images/source',
        tgt_dir='test_images/target',
        output_dir='faceadapter_results'
    )
```

```python
# torch version issue
with open('/content/Face-Adapter/third_party/model_parsing.py', 'r') as f:
    content = f.read()

# replace torch.load without weights_only
content = content.replace(
    "state_dict = torch.load('./checkpoints/third_party/resnet18-5c106cde.pth')",
    "state_dict = torch.load('./checkpoints/third_party/resnet18-5c106cde.pth', weights_only=False)"
)

content = content.replace(
    "state_dict = modelzoo.load_url(resnet18_url)",
    "state_dict = modelzoo.load_url(resnet18_url, weights_only=False)"
)

with open('/content/Face-Adapter/third_party/model_parsing.py', 'w') as f:
    f.write(content)
```
